```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Exemples triviaux pour verifier le kmean

```python
# Importing the dataset
dataset = pd.read_csv('basic4groups.csv')

# plot the data with matplotlib, axis are x0, x1 and points are colored by cluster
plt.scatter(dataset['x0'], dataset['x1'], c=dataset['cluster'], cmap='rainbow')
plt.show()

```

```python
# Importing the dataset
dataset = pd.read_csv('gmm2d.csv')

# plot the data with matplotlib, axis are x0, x1 and points are colored by cluster
plt.scatter(dataset['x0'], dataset['x1'], c=dataset['cluster'], cmap='rainbow')
plt.show()
```

```python
# Importing the dataset
dataset = pd.read_csv('gmm3d.csv')

# plot the data with matplotlib, axis are x0, x1, x2 and points are colored by cluster
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(dataset['x0'], dataset['x1'], dataset['x2'], c=dataset['cluster'], cmap='ocean')
plt.show()
```

```python
dataset = pd.read_csv('predicted.csv')  

# plot the data with matplotlib, axis are idImage,patchId,clusterId
# points are colored by cluster

fig = plt.figure()
ax = fig.add_subplot(111)
# add legend
ax.legend()
ax.set_xlabel('patchId')
ax.set_ylabel('idImage')
ax.scatter(dataset['patchId'], dataset['idImage'], c=dataset['clusterId'], cmap='rainbow')
plt.show()

```
